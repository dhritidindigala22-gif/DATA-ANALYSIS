```python
import pandas as pd
# LOADING DATA 
df=pd.read_csv("social.csv.csv")
# Data Exploration
print(df.head())
```

```python
df.info()
```

```python
SUM OF NULL VALUES IN EACH COLUMN:
```

```python
df.isnull().sum()
```

```python
FILLING NULL VALUES WITH MEAN IN LIKES COLUMN
```

```python
mean_col1=df['likes'].mean()
df.fillna({'likes':mean_col1},inplace=True)
print("after filling null values\n",df,"\n")
```

```python
df.head(10)
```

```python
FILLING NULL VALUES WITH MEAN IN COMMENTS COLUMN
```

```python
mean_col1=df['comments'].mean()
df.fillna({'comments':mean_col1},inplace=True)
print("after filling null values\n",df,"\n")
```

```python
df.head(10)
```

```python
FILLING NULL VALUES WITH MEDIAN IN SHARES COLUMN
```

```python
df.fillna(df['shares'].median(),inplace=True)
print("after filling null values\n",df,"\n")
```

```python
df.head(10)
```

```python
FILLING NULL VALUES WITH MODE IN POSTED_ON COLUMN
```

```python
df['Posted_On']=df['Posted_On'].fillna(df['Posted_On'].mode()[0])
df.fillna({'Posted_On':df['Posted_On'].mode()[0]},inplace=True)
print("after filling null values\n",df,"\n")
df.head(10)
```

```python
df.isnull().sum()
```

```python
valid_days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
df = df[df['Posted_On'].isin(valid_days)]
df['Posted_On'] = df['Posted_On'].astype(str)
df['Posted_On'].unique()
print(df['Posted_On'].mode()[0])
```

```python
df.head(10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("social.csv.csv")

df['post_time'] = pd.to_datetime(df['post_time'], errors='coerce')
df['posted_on'] = df['post_time'].dt.day_name()

df['engagement'] = df['likes'] + df['comments'] + df['shares']

day_eng = df.groupby('posted_on')['engagement'].mean().sort_values(ascending=False)

print(day_eng)

day_eng.plot(kind='bar', title="Best Day to Post")
plt.show()
```

```python
combo = df.groupby(['platform', 'content type'])['engagement'].mean()
print(combo)
combo.unstack().plot(kind='bar', title="Engagement by Platform and Content Type")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("social.csv.csv")
df['engagement'] = df['likes'] + df['comments'] + df['shares']

roi_platform = df.groupby('platform')['engagement'].mean().sort_values(ascending=False)

print(roi_platform)
roi_platform.plot(kind='bar', title="ROI (Engagement) per Platform")
plt.ylabel("Average Engagement")
plt.show()
```
